```python
# Exploratory Data Analysis [EDA]
# Exploratory Data Analysis refers to the critical process of
# performing initial investigations on data so as to discover patterns,
# to spot anomalies,to test hypothesis and to check assumptions
# with the help of summary statistics and graphical representations.

# It is a good practice to understand the data first and try to gather
# as many insights from it.

# Exploratory data analysis (EDA) is used by data scientists to analyze
# and investigate data sets and summarize their main characteristics,
# often employing data visualization methods.

# The main purpose of EDA is to help look at data before making any
# assumptions. It can help identify obvious errors, as well as better
# understand patterns within the data, detect outliers or
# anomalous events, find interesting relations among the variables.

# DATASET : https://raw.githubusercontent.com/harshbg/Telecom-Churn-Data-Analysis/master/Telecom%20Churn.csv
# First Step -- Provided Dataset
```

```python
# how much on average, active user spend on phone on daytime
print('Active Users on Average Spend This Much on Day Time : ',
      data[data['churn']==0]['total day minutes'].mean())

# how much on average, churn user spend on phone on daytime
print('Churn Users on Average Spend This Much on Day Time : ',
      data[data['churn']==1]['total day minutes'].mean())
```

```python
# how much on average, active user spend on phone during evening time
print('Active Users on Average Spend This Much on Evening Time : ',
      data[data['churn']==0]['total eve minutes'].mean())

# how much on average, churn user spend on phone during evening time
print('Churn Users on Average Spend This Much on Evening Time  : ',
      data[data['churn']==1]['total eve minutes'].mean())

# how much on average, active user spend on phone during night time
print('Active Users on Average Spend This Much on Evening Time : ',
      data[data['churn']==0]['total night minutes'].mean())

# how much on average, churn user spend on phone during night time
print('Churn Users on Average Spend This Much on Evening Time  : ',
      data[data['churn']==1]['total night minutes'].mean())
```

```python
# get the average values for churn users
data[data['churn']==1].mean(numeric_only=True)
```

```python
# get the average values for active users
data[data['churn']==0].mean(numeric_only=True)
```

```python
# Grouping
columns_to_show = ['total day minutes','total eve minutes','total night minutes','total intl minutes']
data.groupby('churn')[columns_to_show].mean()
```

```python
# agg() function in pandas let you apply one or more summary function
# perform multiple statistic at once

data.groupby('churn')[columns_to_show].agg(['mean','std','min','max'])
```

```python
columns_to_show = ['total day minutes','total eve minutes','total night minutes','total intl minutes']
data.groupby(['churn'])[columns_to_show].agg([np.mean,np.min,np.max])
```
