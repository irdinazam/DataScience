```python
# Pandas Data Science Library

import numpy as np
import pandas as pd

print(pd.__version__)
```

```python
# Numpy Array
arr = np.array([10, 20, 30, 40, 50])
print(arr)
print(type(arr))
```

```python
# Pandas Series
# A Series is 1D Labeled Array

s = pd.Series([10,20,30,40])
print(type(s))
print(s)
print(s.shape)

# in NumPy -> only Numbers
# in Pandas -> Numbers + (Labels)Index
```

```python
info = pd.Series(['irdina',20,43000,'Bangi'],index=['name','age','postcode','city'])
print(info)
print(info['name'])
print(info['postcode'])
print(info[2])    # will give you warning
print(info.iloc[2])

# Series is like Dictionary + Array combo
```

```python
# values
print(info.values)
```

```python
# Indexing Slicing and Using Loops
s = pd.Series([10,20,30,40,50,60,70,80],index=['val1','val2','val3','val4','val5','val6','val7','val8'])
print(s)
```

```python
# Indexing in Series can be done by using [] or iloc[] or loc[]
# [] label
# .iloc for both label and indexnum
print(s['val1'])        # [] for labels
print(s.iloc[0])        # .iloc for index number
print(s.loc['val1'])    # .loc for labels
```

```python
# Slicing
# 10,20,30,40,50,60,70,80
# 0  1  2  3  4  5  6  7

# when slicing by indexNum, the end is not included
print(s[0:4])

#  when slicing by labels, the end is included
print(s['val1':'val4'])
```

```python
print(s[['val1','val4','val7']])
```

```python
# conditional selection
print(s[s>25])
print(s[s%3 == 0])
```

```python
# using loops (no index/label)
for i in s:
    print(i)
```

```python
for k,v in s.items():
    print(k,v)
```

```python
for i in enumerate(s):
  print(i)
```

```python
# read_csv
# used to read csv file

df = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
df
```

```python
# finding the sum of salary using Pandas
salaries = df['Salary']
print(type(salaries))

total_salary = salaries.sum()
print(total_salary)
```

```python
# finding the sum of salary using NumPy   |   faster than Pandas
salaries = df['Salary'].to_numpy()
print(type(salaries))

total_salary = np.sum(salaries)
print(total_salary)
```
